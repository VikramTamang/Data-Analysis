â€¢ Basic Statistics:
1. Calculate the mean, median, and standard deviation for the Score column.
2. Identify the country with the highest and lowest happiness scores.

```python
#1.
print("1.The mean, median, and standard deviation for the Score column.\n")
mean_score = df['score'].mean()
median_score = df['score'].median()
std_score = df['score'].std()

print("Mean Score:", mean_score)
print("Median Score:", median_score)
print("Standard Deviation of Score:", std_score)
```

```python
#2.
print("2.The country with the highest and lowest happiness scores.\n")
max_score_Country_name = df.loc[df['score'].idxmax(), 'Country name']
min_score_Country_name = df.loc[df['score'].idxmin(), 'Country name']

print("Country with the highest score:", max_score_Country_name)
print("Country with the lowest score:", min_score_Country_name)
```

â€¢ Filtering and Sorting:
1. Filter the dataset to show only the countries with a Score greater than 7.5.

```python
#1.
filtered_df = df[df['score'] > 7.5]
print("Filtered Dataset:\n")
filtered_df.head(143)
```

â€¢ Adding New Columns:
1. Create a new column called Happiness Category that categorizes countries into three categories
based on their Score:

Low âˆ’ (Score < 4)
Medium âˆ’ (4 â‰¤ Score â‰¤ 6)
High âˆ’ (Score > 6)

```python
# Create a new column 'Happiness Category' based on the 'score' column
conditions = [
    (df['score'] < 4),
    (df['score'] >= 4) & (df['score'] <= 6),
    (df['score'] > 6)
]

# Corresponding categories
categories = ['Low', 'Medium', 'High']

# Create the 'Happiness Category' column
df['Happiness Category'] = np.select(conditions, categories, default='Unknown')

df.head(143)
```

2. Data Visualizations:

â€¢ Bar Plot: Plot the top 10 happiest countries by Score using a bar chart.

â€¢ Line Plot: Plot the top 10 unhappiest countries by Score using a Line chart.

â€¢ Plot a histogram for the Score column to show its distribution and also interpret.

â€¢ Scatter Plot: Plot a scatter plot between GDP per Capita and Score to visualize their relationship.

```python
#Bar Plot: Plot the top 10 happiest countries by Score using a bar chart.
top_ten_happiest = df.sort_values(by='score', ascending=False).head(10)

plt.figure(figsize=(16, 8))
plt.bar(top_ten_happiest['Country name'], top_ten_happiest['score'], color = 'green')
plt.xlabel('Country')
plt.ylabel('Score')
plt.title('Top 10 Happiest Countries by Score')

# Display the plot
plt.gca()
plt.show()
```

```python
#Line Plot: Plot the top 10 unhappiest countries by Score using a Line chart.
top_ten_unhappiest = df.sort_values(by='score', ascending=True).head(10)

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(top_ten_unhappiest['Country name'], top_ten_unhappiest['score'], marker='o', color='orange', linestyle='-', linewidth=2)

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.title('Top 10 Unhappiest Countries by Score')

# Display the plot
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Plot a histogram for the Score column to show its distribution and also interpret.

# Create the histogram for the 'score' column
plt.figure(figsize=(12, 6))
plt.hist(df['score'], bins=15, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Happiness Score')
plt.ylabel('Number of Countries')
plt.title('Distribution of Happiness Scores')

# Display the plot
plt.show()
```

```python
#Scatter Plot: Plot a scatter plot between GDP per Capita and Score to visualize their relationship.

plt.figure(figsize=(12, 6))
plt.scatter(df['Log GDP per capita'], df['score'], color='purple', alpha=0.5)

# Add labels and title
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Relationship between Log GDP per Capita and Happiness Score')

# Display the plot
plt.show()
```
